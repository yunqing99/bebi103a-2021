*Explain in words what each of the following code cells does as we work toward tidying this data frame.*



```python
import pandas as pd
import os
# set up the path
data_path = "../../data/"

# import csv file as df
df = pd.read_csv(os.path.join(data_path, "penguins_subset.csv"), header=[0, 1])

df.head()
```

- "df.columns.names = ['species', None]" 

  This names the columns, which add a title to specify that Gentoo, Adelie, and Chinstrap are trhee different speices. 


- "df = df.stack(level='species')" 

  This takes a wide data frame and makes it narrower by converting the names at a level of a columnar multiindex into an index in the data frame.Specifically, for each experiment index, we have will measure the same four properties for three species, so this helps us to stack the four properties together and give index by their species. 
  
  
- "df = df.reset_index(level='species')"

  This makes a new index, based on species, into a column.
  
  
- "df = df.reset_index(drop=True)"

  This convert the row index to columns of the data frame. And when a new index is obtained, the original index is dropped, not retained.
