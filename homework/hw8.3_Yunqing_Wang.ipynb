#### b) If possible, derive an analytical expression for the maximum likelihood estimate of $ \theta$. If you have the expression, compute the MLE for $ \theta$  for each of the three strains. If not, numerically compute the MLE for $ \theta$ for each of the three strains.

The PDF and the Likelihood function for the Binomial distribution are

$ \begin{array}{l}
f(n ; N, \theta)=\left(\begin{array}{c}
N \\
n
\end{array}\right) \theta^{n}(1-\theta)^{N-n} . \\
L(\theta ; N, n)=\left(\begin{array}{c}
N \\
n
\end{array}\right) \theta^{n}(1-\theta)^{N-n} .
\end{array} $

The loglikelihood function is 

$
\ell(\theta ; N, n)=\ln \left(\begin{array}{c}
N \\
n
\end{array}\right)+n \ln \theta+(N-n) \ln (1-\theta) $

Therefore, $ \theta$ MLE $ \theta ^*$ is where $d\ell(\theta ; N, n) / d \theta = 0$, that is
 $ \theta ^* = n /N $

For each strains, 

$ \theta ^*$_WT = 6/50 = 0.120

 $ \theta ^*$_ASH = 15/52 = 0.288
 
 $ \theta ^*$_AVA = 40/53 = 0.755

```python
# use parametric boostrap to calculate the CI
rg = np.random.default_rng()
def bs_ci(n, N, size=1):
    return np.percentile(rg.binomial(N, n / N, size=size) / N, [2.5, 97.5])
```

```python
# calculate each strain's ci
WT_theta_ci = bs_ci(6, 50, size=1000)
ASH_theta_ci = bs_ci(15, 52, size=1000)
AVA_theta_ci = bs_ci(40, 53, size=1000)
```
